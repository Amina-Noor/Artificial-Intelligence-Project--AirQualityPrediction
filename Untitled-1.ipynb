```python
# air_quality_visuals.py

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# ---------------------------
# Load Dataset
# ---------------------------
df = pd.read_csv("air_quality_200k.csv")

# Preview dataset
print(df.head())

# ---------------------------
# 1. Histogram of PM10
# ---------------------------
plt.figure(figsize=(8,5))
sns.histplot(df['PM10'], bins=50, kde=True, color='skyblue')
plt.title("Distribution of PM10 Levels")
plt.xlabel("PM10 (μg/m³)")
plt.ylabel("Frequency")
plt.savefig("PM10_histogram.png", dpi=300)
plt.show()

# ---------------------------
# 2. Time-series Plot of PM10
# ---------------------------
# Convert 'Date' to datetime if not already
df['Date'] = pd.to_datetime(df['Date'])
df_sorted = df.sort_values('Date')

plt.figure(figsize=(12,6))
plt.plot(df_sorted['Date'], df_sorted['PM10'], color='orange')
plt.title("Time-Series of PM10 Levels")
plt.xlabel("Date")
plt.ylabel("PM10 (μg/m³)")
plt.savefig("PM10_timeseries.png", dpi=300)
plt.show()

# ---------------------------
# 3. Correlation Heatmap
# ---------------------------
numeric_features = ['PM10', 'PM2.5', 'CO', 'NO2', 'SO2']
plt.figure(figsize=(8,6))
sns.heatmap(df[numeric_features].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Pollutants")
plt.savefig("pollutant_correlation.png", dpi=300)
plt.show()

# ---------------------------
# 4. PCA Explained Variance Plot
# ---------------------------
X = df[numeric_features].dropna()  # drop missing for PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA()
pca.fit(X_scaled)

plt.figure(figsize=(8,5))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_, marker='o')
plt.title("PCA Explained Variance Ratio")
plt.xlabel("Principal Component")
plt.ylabel("Variance Ratio")
plt.grid(True)
plt.savefig("pca_variance.png", dpi=300)
plt.show()

# ---------------------------
# 5. Actual vs Predicted PM10 (Dummy Example)
# Replace y_pred with your model predictions
# ---------------------------
# Example: Random Forest predictions
y_true = df['PM10'].dropna().values[:1000]  # just taking first 1000 for plotting
import numpy as np
y_pred = y_true + np.random.normal(0, 20, size=len(y_true))  # dummy predictions

plt.figure(figsize=(8,6))
plt.scatter(y_true, y_pred, alpha=0.5, color='green')
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')
plt.title("Actual vs Predicted PM10")
plt.xlabel("Actual PM10")
plt.ylabel("Predicted PM10")
plt.savefig("actual_vs_predicted.png", dpi=300)
plt.show()
```
